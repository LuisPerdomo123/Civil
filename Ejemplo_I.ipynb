# Calculo de fuerzas internas de una viga usando NumPy



El objetivo de este ejemplo es calcular el momento flector y el corte de viga simplemente apoyada usando NumPy. La viga se muestra en la siguiente figura:

image.png



Donde:

q =  20 kN/m

l=5m

En la siguiente línea de código, se almacena la longitud de la viga y el valor de la carga distribuida en dos variables:

```python
import numpy as np

l=5 #m
q = 20 #kN/m
```

## Cálculo del momento flector y corte de viga


Las reacciones del suelo y el sistema de coordenadas del haz se muestran en la siguiente cifra:

image.png

Las ecuaciones de equilibrio son:

image.png

Resolviendo el sistema se obtiene

image.png

Por tanto, el momento flector **M** y el corte de viga **V** son iguales a:

image.png


```python
x = np.linspace(0,l,20)

M = q/2*(l*x-x**2)
V = q*(l/2-x)
```

En la línea 1 creamos la coordenada x usando la función NumPy linspace. Como se observa pasamos l como valor final y 20 como número de pasos, lo que significa que NumPy creará un vector de números espaciados uniformemente que van de 0 a l, que en este caso es igual a 5. A partir de ahora es sólo cuestión de copiar las fórmulas: NumPy realizará todas las operaciones por elementos, entonces M y V serán matrices compuestas por 20 elementos cada una, correspondientes a cada valor de coordenada contenido en x.

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,4))
plt.plot([0]*len(x), color='k')
plt.plot(-M)
plt.plot(V)
```
